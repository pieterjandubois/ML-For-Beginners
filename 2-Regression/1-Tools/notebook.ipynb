```python
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model, model_selection

# Load the Linnerud dataset (X = exercises, y = physiological)
X, y = datasets.load_linnerud(return_X_y=True)

# Select one feature (Situps = column 1 from X) and one target (Waist = column 1 from y)
X_situps = X[:, 1].reshape(-1, 1)  # Situps
y_waist = y[:, 1]                  # Waistline

# Split into train/test sets
X_train, X_test, y_train, y_test = model_selection.train_test_split(
    X_situps, y_waist, test_size=0.33
)

# Create and fit linear regression model
model = linear_model.LinearRegression()
model.fit(X_train, y_train)

# Predict values for test data
y_pred = model.predict(X_test)

# Plot scatter of test data and regression line
plt.scatter(X_test, y_test, color="black", label="Actual data")
plt.plot(X_test, y_pred, color="blue", linewidth=2, label="Regression line")
plt.xlabel("Situps")
plt.ylabel("Waistline")
plt.title("Relationship Between Situps and Waistline")
plt.legend()
plt.show()

print("First 5 exercise rows (X):")
print(X[:5])   # Chins, Situps, Jumps

print("First 5 physiology rows (y):")
print(y[:5])   # Weight, Waist, Pulse
```
